```python
from IPython.display import display 
import os
# Note that Spot (loaded by the kernel) will store a copy of
# the environment variables the first time it reads them, so
# if you change those variables, the new values will be ignored
# until you restart the kernel.
os.environ['SPOT_DOTEXTRA'] = 'size="10.2,5" node[style=filled,fillcolor="#ffffaa"] edge[arrowhead=vee, arrowsize=.7]'
os.environ['SPOT_DOTDEFAULT'] = 'rbcf(Lato)'
import spot
```

# Reading automata output from processes

If an argument of `spot.automata` ends with `|`, then it is interpreted as a shell command that outputs and automaton.

```python
for a in spot.automata('ltl2tgba -s "a U b"; ltl2tgba --lbtt "b"|', 'ltl2tgba -H "GFa" "a & GFb"|'):
    display(a)
```

A single automaton can be read using `spot.automaton()`, with the same convention.

```python
spot.automaton('ltl2tgba -s6 "a U b"|')
```

# Error handling

If the shell command terminates with a non-zero exit status, we should get an exception.

```python
spot.automaton('non-existing-command|')
```

```python
for a in spot.automata("ltl2tgba -H 'a U b'|", 'ltl2tgba -f "syntax U U error"|'):
    display(a)
```

Reading an empty file with `spot.automaton()` is an error.

```python
spot.automaton('true|')
```

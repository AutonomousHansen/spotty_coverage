```python
from pylab import *
import matplotlib.pyplot as plt
import pandas
from mpl_toolkits.mplot3d import Axes3D
import os
```

```python
colors = {
    1: 'grey',  # sl(a) x sl(!a)
    2: 'green', # sl(cl(a)) x !a
    3: 'blue', # cl(sl(a)) x !a
    4: 'orange', # sl2(a) x sl2(!a)
    5: 'purple', # sl2(cl(a)) x !a
    6: 'black', # cl(sl2(a)) x !a
    7: 'red', # on-the-fly sl(a) x sl(!a)
    8: 'magenta', # cl(a) x cl(!a)
}
labels = {
    1: 'sl(a) x sl(!a)',
    2: 'sl(cl(a)) x !a',
    3: 'cl(sl(a)) x !a',
    4: 'sl2(a) x sl2(!a)',
    5: 'sl2(cl(a)) x !a',
    6: 'cl(sl2(a)) x !a',
    7: 'on-the-fly sl(a) x sl(!a)',
    8: 'cl(a) x cl(!a)'
}
```

### By number of atomic propositions
Plots the time taken for each algorithm to check for stutter-invariance, using a csv file generated by "bench/stutter/stutter_bench.sh"

Take the mean among n formulas, where n is at least samples[ap]

```python
x = 'states'
y = 'time'
aps = [1, 2, 3, 4, 5, 6, 7, 8]
samples = {
    1 : 40,
    2 : 40,
    3 : 30,
    4 : 20
}

df = pandas.read_csv('bench_formulas.csv',\
                     names=['formula', 'algo', 'ap', 'states', 'result', 'time'], header=0)
fig = figure(figsize=(20, 10))
for ap in [1, 2, 3, 4]:
    subplot(2, 2, ap)
    title(str(ap) + ' ap')
    df_ap = df[(df['ap'] == ap)]
    for algo in aps:
        a = df_ap[(df_ap['algo'] == algo)][[x, y]]
        size_x = a.groupby(x, as_index=False).size()
        x_gtsamples = [i for i in a[x] if size_x[i] >= samples[ap]]
        a = a[(a[x].isin(x_gtsamples))]
        
        ylabel(y)
        xlabel(x)
        mean = a.groupby(x, as_index=False).mean()
        plot(mean[x], mean[y], color=colors[algo])
        
legends = [matplotlib.lines.Line2D([], [], color=colors[ap], 
                          markersize=15, label=labels[ap]) for ap in aps]
plt.legend(bbox_to_anchor=(1, 3), handles=legends)
```

### By number of atomic propositions and by result

```python
x = 'states'
y = 'time'
aps = [1, 2, 3, 4, 5, 6, 7, 8]
samples = {
    1 : 20,
    2 : 20,
    3 : 20,
    4 : 20
}

df = pandas.read_csv('/home/zehir/prog/spot/bench/stutter/bench_formulas.csv',\
                     names=['formula', 'algo', 'ap', 'states', 'result', 'time'], header=0)
for result in [0, 1]:
    fig = figure(figsize=(20, 10))

    fig.suptitle('stutter-invariant: ' + ('yes' if result == 1 else 'no'))
    for ap in [1, 2, 3, 4]:
        subplot(2, 2, ap)
        title(str(ap) + ' ap')
        df_ap = df[(df['ap'] == ap)]
        for algo in [1, 2, 3, 4, 5, 6, 7, 8]:
            a = df_ap[(df_ap['algo'] == algo)][[x, y]]
            ar = df_ap[(df_ap['algo'] == algo) & (df_ap['result'] == result)][[x, y]]
            size_x = ar.groupby(x, as_index=False).size()
            x_gtsamples = [i for i in ar[x] if size_x[i] >= samples[ap]]
            ar = ar[(ar[x].isin(x_gtsamples))]

            ylabel(y)
            xlabel(x)
            mean = ar.groupby(x, as_index=False).mean()
            plot(mean[x], mean[y], color=colors[algo])

    legends = [matplotlib.lines.Line2D([], [], color=colors[ap], 
                              markersize=15, label=labels[ap]) for ap in aps]
    plt.legend(bbox_to_anchor=(1, 3), handles=legends)
```

### Historgrams

```python
x = range(1, 41)
df = pandas.read_csv('bench_formulas.csv',\
                     names=['algo', 'ap', 'states', 'result', 'time'], header=0)
for ap in [1, 2, 3, 4]:
    a0 = df[(df['ap'] == ap) & (df['result'] == 0) & (df['algo'] == 1)].groupby('states', as_index=False).size()
    a1 = df[(df['ap'] == ap) & (df['result'] == 1) & (df['algo'] == 1)].groupby('states', as_index=False).size()
    y0, y1 = [], []
    for i in x:
        try:
            y0.append(a0[i])
        except KeyError:
            y0.append(0)
        try:
            y1.append(a1[i])
        except KeyError:
            y1.append(0)

    p1 = bar(x, y1)
    p0 = bar(x, y0, bottom=y1, color='y')
    xlabel('states')
    ylabel('number of formulas')
    legend((p1, p0), ('stutter-insensitive', 'stutter-sensitive'))
    show()
```
